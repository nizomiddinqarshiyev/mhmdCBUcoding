```python
# 1️⃣ Bo'sh qiymatlarni 0 bilan to'ldirish (agar kerak bo'lsa)
df['loan_amount'] = df['loan_amount'].fillna('0').astype(str)

# 2️⃣ $ belgilarini olib tashlash
df['loan_amount'] = df['loan_amount'].str.replace('$', '', regex=False)

# 3️⃣ float ga o'tkazish
df['loan_amount'] = df['loan_amount'].astype(float)

# 4️⃣ Tekshirish
print(df['loan_amount'].head())
```

```python
# loan_type ustunini stringga o'tkazish va bo'sh joylarni olib tashlash
df['loan_type'] = df['loan_type'].astype(str).str.strip().str.lower()

# Standartlashtirish uchun map yaratish
loan_type_map = {
    'personal': 'Personal',
    'personal loan': 'Personal',
    'mortgage': 'Mortgage',
    'home loan': 'Mortgage',
    'credit card': 'Credit Card',
    'cc': 'Credit Card'
}

# Qo'llash
df['loan_type'] = df['loan_type'].map(loan_type_map)

# Natijani tekshirish
print(df['loan_type'].unique())
```
